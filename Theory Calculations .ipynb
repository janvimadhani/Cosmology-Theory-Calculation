## Ratio between Temperature Fluctuation and Polarization Correlation Function's Monopoles

https://doi.org/10.1103/PhysRevD.52.3276

Generalized ratio (1):
$$ \frac {C_l^T}{C_l^P} = \bigg(\frac{r^2\,(1+R)}{\Delta\tau_d\, \tau_d} \bigg)^2 \big( 1 + \frac{g(\Omega)}{l}\big) B_l$$
$ $
For flat universe, this simplifies to (2):
$$ \frac {C_l^T}{C_l^P} = N_l\: G\,(\Omega_0,l)$$
$$ \frac {C_l^T}{C_l^P} = N_l \big[ 1 + \frac{0.33 [(1-\Omega_0)/\,\Omega_0]^{2.33}}{l}\big] \Omega_0^{0.22}$$

### Replicate simplified Equation 2 for Flat Universe

```python
#Let l = 30, to replicate paper
l = 30
G = lambda omega_0: (1 + ((0.33) * ((1-omega_0)/omega_0)**(2.33))/l) * omega_0 **(0.22)

omega = np.linspace(0.01,1,100)
plt.figure(figsize=(10,10))
#plt.grid()
plt.plot(omega,G(omega),'k-')
plt.ylim(0.6,2)
plt.xlabel('$\Omega_0$')
plt.ylabel('$G(\Omega_0,l=30)$')
plt.text(0.6,1.6, '$\Omega_0 + \Omega_{\Lambda} = 1, l = 30$',fontsize=16)
plt.show()
```

### Solve R($\eta$):

Red shift as a function of scale factor (8):
$$ 1 + z = \frac{a_0}{a} \; \; (a)$$ 
$ $
$$ z = \frac {a_0}{a} - 1$$
Conformal time (9):
$$ d\eta = \int \frac{dt}{a}  \; \; (a) $$
$ $
$$ = \int \frac{da}{a} \frac{1}{H} \frac{a_0}{a}  \; \; (b)$$
<br>
<br>


In order to rewrite the Friedman equations in terms of conformal time $dt= a(\eta)d\eta$, $\frac{d}{dt} =\frac{d\eta}{dt}\frac{d}{d\eta} =\frac{1}{a}\frac{d}{d\eta}$ and therefore $\left(\frac{\dot{a}}{a}\right)^2 =\frac{1}{a^2}\left(\frac{da}{d\eta} \frac{d\eta}{dt}\right)^2 =\frac{(a')^2}{a^4}$ where the prime denotes the derivative with respect to the conformal time $\eta$. Then, get \begin{align*} \ddot{a}=&\frac{d}{dt} \frac{da}{dt}= \frac{d}{dt}\frac{da}{d\eta}\frac{d\eta}{dt} =\frac{d}{dt}\frac{a'}{a}=\\ =&\frac{d\eta}{dt}\frac{d}{d\eta}\left( \frac{a'}{a}\right)= \frac{1}{a}\left(-\frac{a'^2}{a^2}+\frac{a''}{a}\right)= -\frac{a'^2}{a^3}+\frac{a''}{a^2}. \end{align*} After substitution of the derivatives into the Friedman equations: \begin{align*} & (a')^2 +ka^2=\frac{8\pi G}{3}\rho a^4,\\ &a'' +ka=\frac{4\pi G}{3}\left(\rho-3p\right)a^3. \end{align*}
$ $


#### R($\eta$) for Radiation Dominated Universe and Matter Dominated :

1) Assume flat universe, $k$ = 0.

a) Radiation Dominated ($P = 1/3 \rho$ , $a^4\rho = const$):
$$ \frac {\dot a^2}{a^2} = \frac {8 \pi G}{3} \rho_0 \bigg(\frac{a_0}{a}\bigg) ^4 $$
<br>
$$ \Rightarrow  \frac {da}{dt} = \sqrt{\frac{8 \pi G \rho_0 a_0^4}{3}}\frac {1}{a} \Rightarrow  \int a \, da = \frac {a^2}{2} + K = \sqrt{\frac{8 \pi G \rho_0 a_0^4}{3}} t $$ 
<br>

At time of Big Bang, $ t = 0$, $a = 0$, so K = 0, and $\rho_0 = \rho_{cr}$, 
$$ \Rightarrow a(t) = \sqrt{2 H_0 t} $$
<br>
$$ t = \frac{a^2} {2 H_0} $$

Now, solve $\eta$:
$$ \eta = \int \frac{dt}{a} = \bigg(\frac {1}{2 H_0}\bigg)^{1/2} \int \frac {1}{t^{1/2}} dt $$
$$ =  \bigg(\frac {1}{2 H_0}\bigg)^{1/2} 2 t^{1/2} = 2 \frac {1}{{2 H_0}^{1/2}} \frac{a}{{2 H_0}^{1/2}} = \frac{a}{H_0}$$ 
<br>
$$ \eta = \frac{a}{H_0} $$


b) Matter Dominated ($P = 0$ , $a^3\rho = const$):
$$ \frac {\dot a^2}{a^2} = \frac {8 \pi G}{3} \rho_0 \bigg(\frac{a_0}{a}\bigg) ^3 $$
<br>
$$ \Rightarrow  \frac {da}{dt} = \sqrt{\frac{8 \pi G \rho_0 a_0^3}{3}}\frac {1}{a^{1/2}} \Rightarrow \int a^{1/2} da = \frac{2}{3} a^{3/2} + K = \sqrt{\frac{8 \pi G \rho_0 a_0^3}{3}} t$$
<br>
Again, at time of Big Bang, $ t = 0$, $a = 0$, so K = 0, and $\rho_0 = \rho_{cr}$, so,
$$ \Rightarrow a(t) = \big(\frac{3}{2} H_0 t\big)^{2/3} $$
$$ t = \frac{2}{3 H_0} a^{3/2} $$

Now, solve $\eta$:
$$\eta = \int \frac{dt}{a} = \bigg (\frac{2}{3 H_0} \bigg)^{2/3} \int \frac{dt}{t^{2/3}} = \bigg (\frac{2}{3 H_0} \bigg)^{2/3} 3 t^{1/3} = \bigg (\frac{2}{3 H_0} \bigg)^{2/3} 3 \bigg( \frac{2}{3 H_0} \bigg) ^{1/3} a^{1/2} = \frac{2}{H_0}a^{1/2} $$
<br> 
$$ \eta = \frac{2}{H_0}a^{1/2} $$


```python
#Solve R using formula b, (1-f_nu)^-1 = 1.68132
a_eq = 1
time = np.linspace(0,1,1000) #baryon-photon equality happened roughly 1 sec after BB 
eta = np.linspace(0,1,1000)

R_ = lambda a: 1.68132 * (3/4) *(omega_b/omega_0)* (a/a_eq) #or input eta in terms of a 

a_radiation = np.sqrt(2*H0) * time**(1/2)
a_rad_eta = H0*eta
eta_rad_a = a_radiation / H0 

a_matter = ((3/2)*H0* time)**(2/3)
a_matt_eta = ((H0*eta)/2)**2
eta_matt_a = (2/H0)*np.sqrt(a_matter)

a_darkenergy = exp(H0*time)
a_wrong = H0*time

#standard model cosmology
eta_rad_t = (1/np.sqrt(2*H0))*2*np.sqrt(time)
eta_matt_t = (2/3*H0)**(2/3) * 3 * time**(1/3)


plt.figure(figsize=(15,6))
plt.subplot(131)
plt.plot(time,a_radiation,label='Radiation Dominated')
plt.plot(time,a_matter,label='Matter Dominated')
plt.xlabel('$Time [s]$')
plt.ylabel('$a(t)$')
plt.title('Scale Factor Dependence, $a(t)$')
plt.grid()
plt.legend()

plt.subplot(132)
plt.plot(time,eta_rad_t,label='Radiation Dominated')
plt.plot(time,eta_matt_t,label='Matter Dominated')
plt.xlabel('$Time [s]$')
plt.ylabel('$\eta(t)$')
plt.title('Conformal Time Dependence, $\eta(t)$')
plt.grid()
plt.legend()

plt.subplot(133)
plt.plot(eta,a_rad_eta,label='Radiation Dominated')
plt.plot(eta,a_matt_eta,label='Matter Dominated')
plt.xlabel('$\eta \;[s]$')
plt.ylabel('$a(\eta)$')
plt.title('Scale Factor Dependence, $a(\eta)$')
plt.grid()
plt.legend()


plt.tight_layout()
plt.show()

```

## Sound Crossing Horizon

$$ r_s(\eta) = \int_{0} ^ {\eta} c_s d\eta' = \int_{0} ^ {\eta} \frac{1}{\sqrt3} d\eta' = \frac{1}{\sqrt3}\eta$$

* $c_s \simeq 1/\sqrt(3) $ 

For standard cosmology: 

(a) Radiation dominated:
$$ \eta  = \sqrt{\frac{1}{2 H_0}} 2 \sqrt{t} = \sqrt{\frac{2t}{H_0}} $$

$$ r_s = \frac{1}{\sqrt3}\sqrt{\frac{2t}{H_0}}$$

(b) Matter dominated:

$$ \eta = \bigg(\frac{2}{3 H_0} \bigg) ^{2/3}3 t^{1/3} = \bigg( \frac{4}{3} \frac{t}{{H_0}^2} \bigg) ^{1/3}$$

$$ r_s = \frac{1}{\sqrt3}\bigg( \frac{4}{3} \frac{t}{{H_0}^2} \bigg) ^{1/3} $$

(c) Melia's model:

$$ \eta = \int \frac{dt}{H_0 t} = \frac{1}{H_0} ln(t) + C $$

$$ r_s =\frac{1}{\sqrt3} \frac{1}{H_0} ln(t)$$

```python
eta = np.linspace(0,100,1000)
t = np.linspace(0,100,1000)

k = [1,2,3]

plt.figure(figsize=(15,10))
for i in k:
    plt.plot(eta,np.cos(i*eta),label='$\Theta (\eta)$, k=' + str(i))
    plt.plot(t,np.cos(i*np.sqrt(2*t/H0)),label='$\Theta (t)$, k=' + str(i))

plt.title('Acoustic Oscillation k-Modes ~ cos($k \eta$), cos($k t$)')
plt.xlabel('$t/\eta$')
    
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))

rs_rad = np.sqrt(1/3) * np.sqrt((2*t)/H0)
rs_mat = np.sqrt(1/3) * ((4/3)* (t/H0**2))**(1/3)
rs_melia = np.sqrt(1/3) * (1/H0)* np.log(t)

plt.plot(t,rs_rad,label='Radiation Dominated')
plt.plot(t,rs_mat,label='Matter Dominated')
plt.plot(t,rs_melia,label="Melia's Model")
plt.ylabel('$r_s(t)$')
plt.xlabel('t')
plt.title('Sound Horizon')
plt.legend()
plt.show()
```

Start from Saha Eqn:

$$ \frac{1 - X_e}{X_e^2} = \frac {2 \zeta(3)}{\pi^2} \eta  \bigg( \frac{2\pi T}{m_e} \bigg) ^{3/2}e^{\frac{- \varepsilon _0}{T}} = \beta$$

$$ X_e = \frac{1}{2\beta}[-1 + \sqrt{1+4\beta} $$

Then make substitution to go from temperature to redshift or scale factor:
$$ T = T_0(1+z) = \frac{T_0}{a} $$

```python
zeta_3 = 1.2
eta = 6e-10 
m_e = 511000 #eV
eps = 13.6

T = np.linspace(0.01,1,1000)
T0 = 2.725

def X_e(T):
    beta = (2*zeta_3/ np.pi**2) * eta * ((2*np.pi*T /m_e) ** (3/2)) * np.exp(eps/T)
    X_e = (1/(2*beta)) * (-1 + np.sqrt(1 + 4*beta))
    
    return X_e

```

```python
plt.figure(figsize=(12,8))
plt.plot(T, X_e(T),'k--',label='Saha Eqn - Equilibrium Solution')
plt.title('Evolution of Free Electron Fraction')
plt.xlabel('T [K]')
plt.ylabel('$X_e$')
plt.xlim(1,0)

plt.legend()
plt.show()
```
